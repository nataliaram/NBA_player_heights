````python
df = pd.read_csv (r'/content/drive/MyDrive/Colab Notebooks/nba_heights.csv')
df.head(10)
````

El siguiente código es para crear una columna llamada nombre completo es $O(n)$

````python
nombre_completo=[]
nombre=list(df['first_name'])
apellido=list(df['last_name'])
for i in range(len(nombre)):
  nombre_completo.append(str(nombre[i])+' '+str(apellido[i]))

df['nombre completo']=nombre_completo
df.head(5)
````

**Primer desarrollo.**

````python
def app3(vector, nombres,sum):
  requeridos=dict()
  conteo=0
  for i in range(0,len(vector)):
    faltante=sum-vector[i]
    if requeridos.__contains__(faltante):
      for j in requeridos.get(faltante):
        print("(", nombres[i],
              ", ",nombres[j],
              ")", sep="")
      conteo+=1
    #requeridos[vector[i]]={i}
    requeridos.update({vector[i]:{i}.union(requeridos.get(vector[i])) if requeridos.get(vector[i]) else {i}})
  if conteo==0:
    print("No se encontraron coincidencias")
````

````python
vector=list(df['h_in'])
nombres=list(df['nombre completo'])
# Modificar el valor de entrada numérico (Suma total de la altura de los dos jugadores)
app3(vector, nombres, 139)
````

````python
def app4(sum):
  vector1=[]
  vector2=[]
  faltante=sum-np.array(df['h_in'])
  nombres=np.array(df['nombre completo'])
  for i in range(len(vector)):
    if faltante[i] in vector:
      nombres1=df['nombre completo'][df['h_in']==faltante[i]]
      vector1.extend(nombres1)
      vector2.extend([nombres[i]]*len(nombres1))
  parejas=[(nombre1, nombre2) for nombre1,nombre2 in zip(vector1,vector2) if nombre1!=nombre2]
  for nombre1,nombre2 in parejas:
    if (nombre2,nombre1) in parejas:
      parejas.remove((nombre1,nombre2))
  return parejas 
````
